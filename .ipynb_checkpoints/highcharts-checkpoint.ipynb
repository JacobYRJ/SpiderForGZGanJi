```python
import pymongo
from string import punctuation
import plot
import charts
```

```python
client = pymongo.MongoClient('localhost',27017)
gzganji = client['gzganji']
item_info = gzganji['item_infoY']
```

```python
for i in item_info.find():
    area = i['area'].split('-')
    item_info.update({'_id':i['_id']},{'$set':{'area':area}})
```

```python
for i in item_info.find().limit(100):
    print(i['area'])
```

```python
area_list=[]
for i in item_info.find():
    gz = i['area']
    if gz[0] == '广州':
        if len(gz) == 2:
            area_list.append(gz[1])
        else:
            area_list.append('不明')
area_index = list(set(area_list))
print(area_index)
```

```python
post_times = []
for index in area_index:
    post_times.append(area_list.count(index))
print(post_times)
```

```python
def data_gen(types):
    length = 0
    if length <= len(area_index):
        for area , times in zip(area_index, post_times):
            data = {
                'name':area,
                'data':[times],
                'type':types
            }
            yield data
            length += 1
```

```python
for i in data_gen('column'):
    print(i)
```

```python
series = [data for data in data_gen('column')]
charts.plot(series, show='inline', options=dict(title=dict(text='广州各区二手物品发帖量')))
```
