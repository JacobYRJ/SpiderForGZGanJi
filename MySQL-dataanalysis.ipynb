```python
import numpy as np
from pandas import DataFrame,Series
import pandas as pd
from sqlalchemy import create_engine
import pymongo
import pymysql
```

```python
client = pymongo.MongoClient('localhost',27017)
gzzhuan = client['gzzhuan']
item_info = gzzhuan['item_infoY']
guangzhou = gzzhuan['guangzhouY']
url_list = gzzhuan['url_listY']
```

```python
for i in guangzhou.find():
    for j in item_info.find():
        if i['url'] == j['url']:
            guangzhou.update({'_id':i['_id']}, {'$set':{'price':j['price']}})
```

```python
columns=['price','want_person','look_time','area','fenlei','title']
df = DataFrame(list(guangzhou.find()),columns = columns)
```

```python
fenlei = []
for i in url_list.find():
    fenlei.append(i['fenlei'])
fenlei = list(set(fenlei))
print(fenlei)
```

```python
fenlei_zw = ['手机','电脑','美容保健','平板电脑','居家用品','摩托车','家具','文体户外','家电','图书音像','电动车','自行车','数码产品','母婴儿童','服饰','笔记本']
fenlei = dict(zip(fenlei,fenlei_zw))
print(fenlei)
```

```python
for i in url_list.find():
    for j in fenlei.keys():
        if i['fenlei'] == j:
            url_list.update({'_id':i['_id']},{'$set':{'fenlei':fenlei[j]}})
```
