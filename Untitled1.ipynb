```python
from pandas import DataFrame,Series
import pandas as pd
import pymongo
import numpy
```

```python
client = pymongo.MongoClient('localhost',27017)
gzzhuan = client['gzzhuan']
item_info = gzzhuan['item_infoY']
url_list = gzzhuan['url_listY']
```

```python
columnsl=['price','want_person','look_time','title','url']
item = DataFrame(list(item_info.find({},{'_id':0,'price':1,'want_person':1,'look_time':1,'title':1,'url':1})),columns=columnsl)
```

```python
columnst = ['url','fenlei']
url = DataFrame(list(url_list.find({},{'_id':0,'url':1,'fenlei':1})),columns=columnst)
```

```python
data = item.merge(url,on='url',how='left')
```

```python
data = data.drop_duplicates(['url'])
```
