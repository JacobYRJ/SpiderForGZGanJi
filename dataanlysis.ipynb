```python
import pymongo
from pyecharts import Bar,Pie
```

```python
import charts
```

```python
client = pymongo.MongoClient('localhost',27017)
gzzhuan = client['gzzhuan']
guangzhou = gzzhuan['guangzhou']
```

```python
for i in guangzhou.find().limit(10):
    print(i)
```

```python
for i in guangzhou.find():
    look = int(i['look_time'])
    guangzhou.update({'_id':i['_id']},{'$set':{'look_time':look}})
```

```python
pipeline = [
    {'$group':{'_id':'$fenlei','want':{'$sum':'$want_person'},'look_time':{'$sum':'$look_time'}}}
]
```

```python
for i in guangzhou.aggregate(pipeline):
    print(i)
```

```python
zhonglei = []
want = []
look_time = []
for i in guangzhou.aggregate(pipeline):
    zhonglei.append(i['_id'])
    want.append(i['want'])
    look_time.append(i['look_time'])
```

```python
pie = Pie('各分类想要人数占比',title_pos='center')
pie.add('',zhonglei,want,is_label_show=True,legend_top='bottom')
pie.render_notebook()
```

```python
pie = Pie('各分类浏览人数占比',title_pos='center',height=500)
pie.add('',zhonglei,look_time,is_label_show=True,legend_top='bottom',is_random=True,legend_orient='vertical',legend_pos='left')
pie.render_notebook()
```

```python
pipeline2 = [
    {'$group':{'_id':'$fenlei','count':{'$sum':1}}}
]
```

```python
for i in guangzhou.aggregate(pipeline2):
    print(i)
```

```python
pie2 = guangzhou.aggregate(pipeline2)
pie2_z = []
pie2_v = []
for i in pie2:
    pie2_z.append(i['_id'])
    pie2_v.append(i['count'])
pie = Pie('各分类发帖量占比',title_pos='center',height=550)
pie.add('',pie2_z,pie2_v ,is_label_show=True,legend_top='bottom',legend_orient='vertical',legend_pos='left')
pie.render_notebook()
```

```python
pipe3 = [
    {'$match':{'$and':[{'fenlei':'手机'},{'want_person':{'$gt':30}}]}},
    {'$sort':{'want_person':-1}},
]
shouji = {}
for i in guangzhou.aggregate(pipe3):
    name = i['title'].split()[0]
    if not name in shouji.keys():
        shouji[name] = 0
        shouji[name] = shouji[name] + i['want_person']
    else:
        shouji[name] = shouji[name] + i['want_person']
print(shouji)
```

```python
pie = Pie('二手手机想要量占比',title_pos='center',height=450)
pie.add('',[s for s in shouji.keys()] ,[v for v in shouji.values()],is_label_show=True,legend_top='bottom',legend_orient='vertical',legend_pos='left')
pie.render_notebook()
```
