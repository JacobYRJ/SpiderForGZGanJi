```python
import numpy as np
from pandas import DataFrame, Series
import pandas as pd
import pymongo
```

```python
client = pymongo.MongoClient('localhost',27017)
gzzhuan = client['gzzhuan']
guangzhou = gzzhuan['guangzhouY']
```

```python
for i in guangzhou.find().limit(5):
    print(i)
```

```python
for i in guangzhou.find():
    st = ''.join(i['area'])
    guangzhou.update({'_id':i['_id']},{'$set':{'area':st}})
```

```python
columns=['price','want_person','look_time','area','fenlei','title']
data = DataFrame(list(guangzhou.find({},{'_id':0,'price':1,'want_person':1,'look_time':1,'area':1,'fenlei':1,'title':1})),columns=columns)
```
